# Python结合Gephi完成可视化
---

```python
import networkx as nx
import pandas as pd

G = nx.read_weighted_edgelist("./stormofswords.csv", delimiter=",", nodetype=str)
```

```python
degree_result = nx.degree(G)
closeness_result = nx.closeness_centrality(G)
betweenness_result = nx.betweenness_centrality(G)
eigenvector_result = nx.eigenvector_centrality(G, max_iter=10000)
pagerank_result = nx.pagerank(G)
G.remove_edges_from(nx.selfloop_edges(G))
kshell_result = nx.core_number(G)

for node in G.nodes():
    G.nodes[node]['degree'] = degree_result[node]
    G.nodes[node]['closeness'] = closeness_result[node]
    G.nodes[node]['betweenness'] = betweenness_result[node]
    G.nodes[node]['eigenvector'] = eigenvector_result[node]
    G.nodes[node]['pagerank'] = pagerank_result[node]
    G.nodes[node]['kshell'] = kshell_result[node]
```

```python
nx.write_gexf(G, "./stormofswords_visualization.gexf")
```
