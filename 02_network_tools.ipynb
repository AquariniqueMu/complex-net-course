## 1、读取网络文件
实践中网络一般来自于读取网络文件，包括txt、edgelist、gexf、csv、graphhml、net等格式

```python
import networkx as nx
```

```python



# G = nx.read_gexf("./physicians.gexf")
# G = nx.read_graphml("./physicians.graphml")
# G = nx.read_pajek("./physicians.net")
# G = nx.read_adjlist("./physicians.txt")
# G = nx.read_edgelist("./physicians.csv", delimiter=",")
G = nx.read_edgelist("./physicians.edgelist")

# nx.write_gexf(G, "./physicians.gexf")
# nx.write_graphml(G, "./physicians.graphml")
# nx.write_pajek(G, "./physicians.net")
# nx.write_edgelist(G, "./physicians.edgelist")
# nx.write_adjlist(G, "./physicians.txt")
# nx.write_edgelist(G, "./physicians.csv", delimiter=",")
```

```python
nx.draw(G, pos=nx.spring_layout(G),  node_size=80, font_size=8, font_color="black", font_weight="bold", node_color="coral", edge_color="black", width=0.8, alpha=0.6)
```

## 2、子图和连通组件

```python
sub_G_1 = nx.subgraph(G, ['1', '2', '3'])
nx.draw(sub_G_1,with_labels=True, node_size=280, font_size=8, font_color="black", font_weight="bold", node_color="coral", edge_color="black", width=0.8, alpha=0.6)
```

```python
components = nx.connected_components(G)
for c in components:
    print(len(c))
```

```python
max_component = max(nx.connected_components(G), key=len)
nx.draw(nx.subgraph(G, max_component), node_size=80, font_size=8, font_color="black", font_weight="bold", node_color="coral", edge_color="black", width=0.3, alpha=0.6)
```

## 3、合成网络

```python
er = nx.erdos_renyi_graph(100, 0.15)
nx.draw(er, node_size=80, font_size=8, font_color="black", font_weight="bold", node_color="coral", edge_color="black", width=0.3, alpha=0.6)
```

```python
ws = nx.watts_strogatz_graph(100, 4, 0.15)
nx.draw(ws, node_size=80, font_size=8, font_color="black", font_weight="bold", node_color="coral", edge_color="black", width=0.3, alpha=0.6)
```

```python
ba = nx.barabasi_albert_graph(100, 4)
nx.draw(ba, node_size=80, font_size=8, font_color="black", font_weight="bold", node_color="coral", edge_color="black", width=0.3, alpha=0.6)
```
