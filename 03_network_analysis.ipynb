## 2、节点中心性算法

```python
G = nx.read_weighted_edgelist("./stormofswords.csv", delimiter=",", nodetype=str)
```

```python
nx.draw(G, with_labels=True, font_weight='bold', node_size=200, node_color='darkorange', font_size=6, font_color='black', edge_color='gray', width=2, style='dotted', alpha=0.5)
```

```python
list(G.nodes())
```

### 度中心性

```python
nx.degree_centrality(G)['Jon']
```

### 接近中心性

```python
if G.is_directed():
    G = G.reverse()  # create a reversed graph view

path_length = nx.single_source_shortest_path_length

nodes = G.nodes

closeness_dict = {}
for n in nodes:
    sp = path_length(G, n)
    totsp = sum(sp.values())
    len_G = len(G)
    _closeness_centrality = 0.0
    if totsp > 0.0 and len_G > 1:
        _closeness_centrality = (len(sp) - 1.0) / totsp
        # normalize to number of nodes-1 in connected part
        s = (len(sp) - 1.0) / (len_G - 1)
        _closeness_centrality *= s
    closeness_dict[n] = _closeness_centrality

closeness_dict
```

```python
nx.closeness_centrality(G)['Jon']
```

```python
nx.pagerank(G)['Jon']
```

### K-shell中心性

```python
G.remove_edges_from(nx.selfloop_edges(G))
nx.core_number(G)['Jon']
```

```python
import numpy as np
import scipy as sp
import scipy.sparse.linalg  # call as sp.sparse.linalg

if len(G) == 0:
    raise nx.NetworkXPointlessConcept(
        "cannot compute centrality for the null graph"
    )
M = nx.to_scipy_sparse_array(G, nodelist=list(G), weight=None, dtype=float)
_, eigenvector = sp.sparse.linalg.eigs(
    M.T, k=1, which="LR", maxiter=10000, tol=0
)
largest = eigenvector.flatten().real
norm = np.sign(largest.sum()) * sp.linalg.norm(largest)
dict(zip(G, largest / norm))
```

### 特征向量中心性

```python
nx.eigenvector_centrality(G, max_iter=10000)['Jon']
```

### 记录结果

```python
import pandas as pd
df = pd.DataFrame(columns=['Degree', 'Closeness', 'Betweenness', 'PageRank', 'Coreness', 'Eigenvector'], index=list(G.nodes()))


for node in G.nodes():
    df.loc[node] = [nx.degree_centrality(G)[node], nx.closeness_centrality(G)[node], nx.betweenness_centrality(G)[node], nx.pagerank(G)[node], nx.core_number(G)[node], nx.eigenvector_centrality(G, max_iter=10000)[node]]
    
df.to_excel('stormofswords.xlsx')
```

```python
df_rank = pd.DataFrame(columns=['Degree', 'Closeness', 'Betweenness', 'PageRank', 'Coreness', 'Eigenvector'], index=list(G.nodes()))

for column in df.columns:
    df_rank[column] = df[column].rank(ascending=False, method='min')

df_rank.to_excel('stormofswords_rank.xlsx')
```
